```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load and standardize data
X, y = load_iris(return_X_y=True)
X = StandardScaler().fit_transform(X)

# K-Means with 3 clusters (since Iris has 3 classes)
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X)

# Plot clusters (first two features)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("K-Means Clustering on Iris")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("Cluster Centers:\n", kmeans.cluster_centers_)
```

```python
#PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.title("PCA on Iris (2 Components)")
plt.show()
```

```python
#HMM
!pip install hmmlearn
from hmmlearn import hmm

model = hmm.GaussianHMM(n_components=3, covariance_type='diag', n_iter=100)
model.fit(X)
hidden_states = model.predict(X)
plt.scatter(X[:,0], X[:,1], c=hidden_states)
plt.title("HMM Clustered States on Iris")
plt.show()
```

```python
#CART
from sklearn.tree import DecisionTreeClassifier, plot_tree

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X_train, y_train)
print("Accuracy:", tree.score(X_test, y_test))
plot_tree(tree, filled=True)
plt.show()
```

```python
#ENSEMBLE

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=0)
rf.fit(X_train, y_train)
print("Random Forest Accuracy:", rf.score(X_test, y_test))

gb = GradientBoostingClassifier(random_state=0)
gb.fit(X_train, y_train)
print("Gradient Boosting Accuracy:", gb.score(X_test, y_test))
```
